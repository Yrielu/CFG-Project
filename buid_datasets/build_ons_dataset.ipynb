```python
import pandas as pd
```

```python
gdp_data = pd.read_csv('data/ONS/series-130425_gdp.csv')
```

```python
unemployment_data = pd.read_csv('data/ONS/series-150425_unemployment.csv')
```

```python
cpi_data = pd.read_csv('data/ONS/series-130425_cpih.csv')
```

```python
consumer_trends_data = pd.read_csv('data/ONS/series-130425_cthhfce.csv')
```

```python
gdp_data.shape
```

```python
pattern = r'20(0[9]|1[0-9]|2[0-4]) Q[1-4]'

gdp_filtered = gdp_data[gdp_data['Title'].str.contains(pattern, na=False)]

gdp_filtered
```

```python
unemployment_data.shape
```

```python
pattern = r'20(0[9]|1[0-9]|2[0-4]) Q[1-4]'

unemployment_filtered = unemployment_data[unemployment_data['Title'].str.contains(pattern, na=False)]

unemployment_filtered
```

```python
cpi_data.shape
```

```python
pattern = r'20(0[9]|1[0-9]|2[0-4]) Q[1-4]'

cpi_filtered = cpi_data[cpi_data['Title'].str.contains(pattern, na=False)]

cpi_filtered
```

```python
consumer_trends_data.shape
```

```python
pattern = r'20(0[9]|1[0-9]|2[0-4]) Q[1-4]'

consumer_trends_filtered = consumer_trends_data[consumer_trends_data['Title'].str.contains(pattern, na=False)]

consumer_trends_filtered
```

```python
# ons_data = []

ons_data = pd.merge(gdp_filtered, unemployment_filtered, on='Title', how='inner')
ons_data = pd.merge(ons_data, cpi_filtered, on='Title', how='inner')
ons_data = pd.merge(ons_data, consumer_trends_filtered, on='Title', how='inner')
```

```python
ons_data
```

```python
ons_data.rename(columns = {'Title':'Year Quarter', 
                       'Gross Domestic Product: chained volume measures: Seasonally adjusted £m':'Gross Domestic Product (£m)', 
                       'Unemployment rate (aged 16 and over, seasonally adjusted): %':'Unemployment Rate (%)', 
                       'CPIH ANNUAL RATE 00: ALL ITEMS 2015=100': 'CPIH (%)',
                       'Household final consumption expenditure :National concept CVM SA - £m': 'Household Final Consumption Expenditure (£m)'}, inplace=True)
```

```python
ons_data
```

```python
ons_data.to_csv('ons_filtered.csv')
```
